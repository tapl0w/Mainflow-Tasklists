```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes=pd.read_csv(r"D:\Main flow assingments\Python Datase\Python Dataset\\archive (3)\diabetes_data.csv")
```

```python
diabetes.head()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')
```

```python
sns.heatmap(diabetes[['Age', 'BMI', 'FastingBloodSugar', 'HbA1c', 'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides']].corr(), annot=True, cmap='cubehelix')
plt.title('Correlation Heatmap')
plt.show()
```

```python
diabetes['Ethnicity'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Ethnicity Distribution')
plt.ylabel('')  # Optional to hide y-label
plt.show()
```

```python

sns.set_theme()


sns.FacetGrid(diabetes, col="Ethnicity").map(sns.histplot, "Age").set_axis_labels("Age", "Count");
```

```python
sns.boxplot(x='Diagnosis', y='QualityOfLifeScore', data=diabetes)
plt.title('Quality of Life Score by Diagnosis')
plt.show()
```

```python
sns.countplot(x='HeavyMetalsExposure', hue='Diagnosis', data=diabetes)
plt.title('Heavy Metals Exposure vs. Diagnosis')
plt.show()
```

```python
sns.kdeplot(data=diabetes['HbA1c'], shade=True,fill=True)
plt.title('HbA1c Distribution')
plt.show();
```

```python
sns.swarmplot(x='AlcoholConsumption', y='PhysicalActivity', data=diabetes)
plt.title('Alcohol Consumption vs. Physical Activity')
plt.show()
```

```python
sns.countplot(x='Smoking', hue='Diagnosis', data=diabetes,palette='mako')
plt.title('Smoking Status vs. Diabetes Diagnosis')
plt.show()
```

```python
sns.swarmplot(x='Ethnicity', y='FastingBloodSugar', data=diabetes)
plt.title('Swarm Plot')
plt.show()
```
